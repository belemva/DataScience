# Pour les dimensions on prend ceux pliées

groupe = DRPp // DCXp 

description = Order N°

pieces = Modified quantity

length = 'Longueur extérieure replié'

width = Largeur extérieure replié

height = Hauteur extérieure replié

total weight = Charge maxi dans l'emballage  ### A reverifier

stackable = Y 

tiltable = Y

rotable = Y

```python
easycargo_columns=['DRPp']+['Order N°']+['Modified quantity']+['Longueur extérieure']+['Largeur extérieure']+['Hauteur extérieure']+["Charge maxi dans l'emballage"]

easycargo_2019_07_01=delivery_2019_07_01.loc[:,easycargo_columns]
easycargo_2019_07_01.head()
```

```python
easycargo_2019_07_01.loc[:,'stackable'] = 'Y'
easycargo_2019_07_01.loc[:,'tiltable'] = 'Y'
easycargo_2019_07_01.loc[:,'rotable'] = 'Y'
easycargo_2019_07_01.loc[:,'Modified quantity'] = 1
for i in range(len(easycargo_2019_07_01['Longueur extérieure'])):
    easycargo_2019_07_01['Longueur extérieure'][i]=str(easycargo_2019_07_01['Longueur extérieure'][i]).replace(' mm','').replace(' ','')
    easycargo_2019_07_01['Largeur extérieure'][i]=str(easycargo_2019_07_01['Largeur extérieure'][i]).replace(' mm','').replace(' ','')
    easycargo_2019_07_01['Hauteur extérieure'][i]=str(easycargo_2019_07_01['Hauteur extérieure'][i]).replace(' mm','').replace(' ','')
    easycargo_2019_07_01['Charge maxi dans l\'emballage'][i]=str(easycargo_2019_07_01['Charge maxi dans l\'emballage'][i]).replace('kg','').replace(' ','').replace(',','.')

easycargo_2019_07_01['Longueur extérieure'] = easycargo_2019_07_01['Longueur extérieure'].astype(float)
easycargo_2019_07_01['Largeur extérieure'] = easycargo_2019_07_01['Largeur extérieure'].astype(float)
easycargo_2019_07_01['Hauteur extérieure'] = easycargo_2019_07_01['Hauteur extérieure'].astype(float)
easycargo_2019_07_01['Charge maxi dans l\'emballage'] = easycargo_2019_07_01['Charge maxi dans l\'emballage'].astype(float)
```

```python
easycargo_2019_07_01.columns=['Group', 'Description','Pieces','Length (mm)', 'Width (mm)','Height (mm)','Total weight (kg)','Stackable','Tiltable','Rotable']
```
